## Methods

```python
def Orig_Bauer(df):
    columns_to_check = ["Katagiri_Primary", "Visceral", "Nr_Skel_Met", "Path_Fract"]
    
    missing_values = df[columns_to_check].isnull().any().any()
    
    if missing_values:
        return "MV"
    
    else:
        score = 0
        
        if (df["Visceral"] == 0 and
            df["Brain"] == 0):
            score += 1
        
        if df["Path_Fract"] == 0:
            score += 1
            
        if df["Nr_Skel_Met"] == 0:
            score += 1
            
        if (df["Katagiri_Primary"] != "Non-small cell lung cancer with molecularly targeted therapy" and 
            df["Katagiri_Primary"] != "Other lung cancer"):
            score += 1
        
        if (df["Katagiri_Primary"] == "Hormone independent breast cancer" or 
            df["Katagiri_Primary"] == "Hormone dependent breast cancer" or 
            df["Katagiri_Primary"] == "Renal cell carcinoma" or 
            df["Katagiri_Primary"] == "Malignant myeloma" or 
            df["Katagiri_Primary"] == "Malignant lymphoma"):
            score += 1
        
        if score in [4, 5]:
            return score, 87, 73, 50
        elif score in [2, 3]:
            return score, 71, 50, 25
        elif score in [0, 1]:
            return score, 20, 0, 0
            
def Mod_Bauer(df):
    columns_to_check = ["Katagiri_Primary", "Visceral", "Nr_Skel_Met"]
    
    missing_values = df[columns_to_check].isnull().any().any()
    
    if missing_values:
        return "MV"
    
    else:
        score = 0
        
        if (df["Visceral"] == 0 and
            df["Brain"] == 0):
            score += 1
            
        if (df["Katagiri_Primary"] != "Non-small cell lung cancer with molecularly targeted therapy" or 
            df["Katagiri_Primary"] != "Other lung cancer"):
            score += 1
        
        if (df["Katagiri_Primary"] == "Hormone independent breast cancer" or 
            df["Katagiri_Primary"] == "Hormone dependent breast cancer" or 
            df["Katagiri_Primary"] == "Renal cell carcinoma" or 
            df["Katagiri_Primary"] == "Malignant myeloma" or 
            df["Katagiri_Primary"] == "Malignant lymphoma"):
            score += 1
            
        if df["Nr_Skel_Met"] == 0:
            score += 1
            
        if score in [3, 4]:
            return score, 97, 88, 75
        elif score in [2]:
            return score, 82, 65, 45
        elif score in [0, 1]:
            return score, 50, 29, 7.1
          
def Rev_Toku(df):
    columns_to_check = ["KPS", "Nr_Ex_Spin_Foci", "Nr_Spine_Met", "Katagiri_Primary", "Frankel"]
    
    missing_values = df[columns_to_check].isnull().any().any()
    
    if missing_values:
        return "MV"
    
    else:
        score = 0
        
        if (df["KPS"] == "80" or
            df["KPS"] == "90" or
            df["KPS"] == "95" or
            df["KPS"] == "100"):
            score += 2
        if (df["KPS"] == "50" or
              df["KPS"] == "55" or
              df["KPS"] == "60" or
              df["KPS"] == "70"):
            score += 1 
        
        if df["Nr_Ex_Spin_Foci"] == 0:
            score += 2
        if df["Nr_Ex_Spin_Foci"] == 1:
            score += 1
            
        if df["Nr_Spine_Met"] == 0:
            score += 2
        if df["Nr_Spine_Met"] == 1:
            score += 1
        
        if df["Removable"] == "removable":
            score += 1
        if df["Removable"] == "No visceral mets":
            score += 2        
        
        if (df["Katagiri_Primary"] == "Thyroid cancer" or
            df["Katagiri_Primary"] == "Hormone dependent breast cancer" or
            df["Katagiri_Primary"] == "Hormone independent breast cancer" or
            df["Katagiri_Primary"] == "Hormone dependent prostate cancer" or
            df["Katagiri_Primary"] == "Hormone independent prostate cancer"):
            score += 5
        if df["Primary Tumor Extra"] == "Rectal":
            score += 4
        if (df["Katagiri_Primary"] == "Renal cell carcinoma" or
              df["Katagiri_Primary"] == "Other gynecological cancer"):
            score += 3
        if (df["Katagiri_Primary"] == "Others" or
              df["Katagiri_Primary"] == "Malignant lymphoma" or
              df["Katagiri_Primary"] == "Malignant melanoma" or
              df["Katagiri_Primary"] == "Malignant myeloma" or
              df["Katagiri_Primary"] == "Colon and rectal cancer" or
              df["Katagiri_Primary"] == "Head and neck cancer" or
              df["Katagiri_Primary"] == "Other urological cancer" or
              df["Katagiri_Primary"] == "Cervical cancer"):
            score += 2
        if (df["Katagiri_Primary"] == "Hepatocellular carcinoma" or
              df["Katagiri_Primary"] == "Gallbladder cancer" or
              df["Katagiri_Primary"] == "Unknown origin"):
            score += 1
        if (df["Frankel"] == 2 or
            df["Frankel"] == 3):
            score += 1
        if df["Frankel"] == 4:
             score += 2 
        
        if score in [12, 13, 14, 15]:
            return score, 100, 100, 91
        elif score in [9, 10, 11]:
            return score, 85, 73, 30
        elif score in [0, 1, 2, 3, 4, 5, 6, 7, 8]:
            return score, 17, 15, 4.5 

def Tomita(df):
    columns_to_check = ["Tomita_Group", "Visceral", "Nr_Skel_Met"]
    
    missing_values = df[columns_to_check].isnull().any().any()
    
    if missing_values:
        return "MV"
    
    else:
        score = 0
        
        if (df["Tomita_Group"] == 1):
            score += 1
        if (df["Tomita_Group"] == 2):
            score += 2
        if (df["Tomita_Group"] == 3):
            score += 4
        
        if df["Treatable"] == "No visceral mets":
            score += 0
        if df["Treatable"] == "treatable":
            score += 2
        if df["Treatable"] == "untreatable":
            score += 4
    
        if (df["Nr_Skel_Met"] == 0):
            score += 1
        if (df["Nr_Skel_Met"] == 1):
            score += 2
            
        if score in [2, 3]:
            return score, 100, 100, 100
        elif score in [4, 5]:
            return score, 100, 100, 100
        elif score in [6, 7]:
            return score, 100, 100, 0
        elif score in [8, 9, 10]:
            return score, 0, 0, 0

def Bollen(df):
    columns_to_check = ["Tomita_Group", "KPS", "Visceral", "Brain"]
    
    missing_values = df[columns_to_check].isnull().any().any()
    
    if missing_values:
        return "MV"
    
    else:
        if df["Clinical Profile"] == "Favorable":
            if (df["KPS"] == "80" or 
                df["KPS"] == "90" or
                df["KPS"] == "95" or
                df["KPS"] == "100"):
                if df["Brain"] == 0 and df["Visceral"] == 0:
                    profile = "A"
                else:
                    profile = "B"
            else:
                if df["Brain"] == 0 and df["Visceral"] == 0:
                    profile = "B"
                else:
                    profile = "C"
                
        if df["Clinical Profile"] == "Moderate":
            if (df["KPS"] == "80" or 
                df["KPS"] == "90" or
                df["KPS"] == "95" or
                df["KPS"] == "100"):
                profile = "B"
            else:
                profile = "C"
            
        if df["Clinical Profile"] == "Unfavorable":
            if (df["KPS"] == "80" or 
                df["KPS"] == "90" or
                df["KPS"] == "95" or
                df["KPS"] == "100"):
                profile = "C"
            else:
                profile = "D"
                
        if profile == "A":
            return profile, 96, 84, 78
        elif profile == "B":
            return profile, 85, 71, 58
        elif profile == "C":
            return profile, 60, 40, 17
        elif profile == "D":
            return profile, 27, 14, 4.5

def van_der_Linden(df):
    columns_to_check = ["KPS","Katagiri_Primary", "Visceral"]
    
    missing_values = df[columns_to_check].isnull().any().any()
    
    if missing_values:
        return "MV"
    
    else:
        score = 0
        
        if (df["KPS"] == "80" or 
            df["KPS"] == "90" or
            df["KPS"] == "95" or
            df["KPS"] == "100"):
            score += 2
        if (df["KPS"] == "50" or
              df["KPS"] == "55" or
              df["KPS"] == "60" or
              df["KPS"] == "70"):
            score += 1
            
        if (df["Katagiri_Primary"] == "Hormone independent breast cancer" or 
            df["Katagiri_Primary"] == "Hormone dependent breast cancer"):
            score += 3
        if (df["Katagiri_Primary"] == "Hormone independent prostate cancer" or 
              df["Katagiri_Primary"] == "Hormone dependent prostate cancer"):
            score += 2   
        if (df["Katagiri_Primary"] == "Non-small cell lung cancer with molecularly targeted therapy" or 
              df["Katagiri_Primary"] == "Other lung cancer"):
            score += 1
            
        if (df["Visceral"] == 0 and
            df["Brain"] == 0):
            score += 1
        
        if score in [0, 1, 2, 3]:
            return score, 50, 27, 6.3
        elif score in [4, 5]:
            return score, 87, 66, 44
        elif score in [6]:
            return score, 90, 83, 60
    
def OSRI(df):
    columns_to_check = ["KPS", "Tomita_Group","Katagiri_Primary", "Visceral"]
    missing_values = df[columns_to_check].isnull().any().any()
    
    if missing_values:
        return "MV"
    
    else:
        score = 0
        
        if (df["Katagiri_Primary"] == "Other lung cancer" or
            df["Katagiri_Primary"] == "Non-small cell lung cancer with molecularly targeted therapy"):
            score += 5
        else:
            if df["Tomita_Group"] == 1:
                score += 1
            if df["Tomita_Group"] == 2:
                score += 2
            if df["Tomita_Group"] == 3:
                score += 4
            
        if (df["KPS"] == "50" or
            df["KPS"] == "55" or
            df["KPS"] == "60" or
            df["KPS"] == "70"):
            score += 1 
        if (df["KPS"] == "40" or
              df["KPS"] == "35" or
              df["KPS"] == "30" or 
              df["KPS"] == "20" or
              df["KPS"] == "15" or  
              df["KPS"] == "10"):
            score += 2
            
        if score in [1]:
            return score, 87, 79, 66
        if score in [2, 3]:
            return score, 62, 49, 37
        if score in [4, 5]:
            return score, 54, 30, 16
        if score in [6]:
            return score, 31, 6.5, 0
        if score in [7]:
            return score, 0, 0, 0

def NESMS(df):
    #Albumine: g/L naar g/dL
    columns_to_check = ["Alb", "Functional_Stat_1"]
    
    missing_values = df[columns_to_check].isnull().any().any()
    
    if missing_values:
        return "MV"
    
    else:
        score = 0
        
        if Mod_Bauer(df)[0] > 2:
            score += 2
            
        if (df["Alb"]/10) >= 3.5:
            score += 1
            
        if df["Functional_Stat_1"] == 1:
            score += 1
            
        if score == 0:
            return score, 41, 30, 19
        elif score == 1:
            return score, 75, 57, 35
        elif score == 2:
            return score, 73, 59, 46
        elif score in [3, 4]:
            return score, 86, 77, 68

def Mizumoto(df):
    df["Calc_cor"] = 0.8 * (4 - (df["Alb"] / 10)) + (df["Calc"] * 4)
    
    columns_to_check = ["Katagiri_Primary", "ECOG", "Visceral", "Pre_Chem",
                        "Calc", "Nr_Skel_Met", "Age"]
    missing_values = df[columns_to_check].isnull().any().any()
    
    if missing_values:
        return "MV"
    
    else:
        score = 0
        
        if (df["Katagiri_Primary"] != "Hormone independent breast cancer" and 
            df["Katagiri_Primary"] != "Hormone dependent breast cancer" and
            df["Katagiri_Primary"] != "Hormone independent prostate cancer" and 
            df["Katagiri_Primary"] != "Hormone dependent prostate cancer" and
            df["Katagiri_Primary"] != "Malignant lymphoma" and
            df["Katagiri_Primary"] != "Thyroid cancer"):
            score += 3
        
        if df["ECOG"] > 2:
            score += 3
        
        if (df["Visceral"] == 1 or
            df["Brain"] == 1):
            score += 2
            
        if df["Pre_Chem"] == 1:
            score += 2
        
        if df["Calc_cor"] > 10:
            score += 2

        if df["Nr_Skel_Met"] == 1:
            score += 1
        
        if df["Age"] > 70:
            score += 1
        
        if score in [0, 1, 2, 3, 4]:
            return score, 99, 89, 77
        elif score in [5, 6, 7, 8, 9]:
            return score, 71, 46, 22
        elif score in [10, 11, 12, 13, 14]:
            return score, 18, 7, 4

def Rev_Kat_individual(df):
    #Albumine: nu g/L, naar g/dL
    #CRP: nu mg/L, naar mg/dL
    #Calcium: nu mmol/L, naar mg/dL, 1 mmol/L = 4.01 mg/dl
    #Bili: nu umol, naar mg/dL, 1umol/l = 0.06 mg/dL
    #Platelets: <100.000/uL (zie artikel, staat fout op internet)
    # is gelijk aan <100 x10^9/L
    columns_to_check = ['Katagiri_Group', 'Visceral', 'Brain', 'Disseminated', 
                        "CRP", "LDH", "Alb", "Platelet", "Calc", "Tot_Bill",
                        "ECOG", "Pre_Chem", "Nr_Skel_Met"]
    
    missing_values = df[columns_to_check].isnull().any().any()
    
    if missing_values:
        return "MV", "MV", "MV"
    
    else:
        score = 0
        
        if df["Katagiri_Group"] == 2:
            score += 2
        if df["Katagiri_Group"] == 3:
            score += 3
            
        if df["Disseminated"] == 1:
            score += 2
        if (df["Disseminated"] == 0 and (df["Visceral"] == 1 or 
                                           df["Brain"] == 1)):
            score += 1
            
        if (df["Platelet"] < 100 or 
            df["Calc"]*4.01 >= 10.3 or 
            df["Tot_Bill"]*0.06 >= 1.4):
            score += 2
        if (df["Platelet"] >= 100 and 
              df["Calc"]*4.01 < 10.3 and
              df["Tot_Bill"]*0.06 < 1.4 and (df["CRP"]/10 >= 4 or 
                                               df["LDH"] > 250 or 
                                               df["Alb"]/10 < 3.7)):
            score += 1
            
        if df["ECOG"] in [3,4]:
            score += 1
            
        if df["Pre_Chem"] == 1:
            score += 1
            
        if df["Nr_Skel_Met"] == 1:
            score +=  1
            
        if score == 0:
            return score, 100, 100, 100
        elif score == 1:
            return score, 100, 100, 95
        elif score == 2:
            return score, 100, 100, 97
        elif score == 3:
            return score, 99, 96, 87
        elif score == 4:
            return score, 95, 95, 75
        elif score == 5:
            return score, 88, 78, 53
        elif score == 6:
            return score, 73, 60, 33
        elif score == 7:
            return score, 52, 40, 10
        elif score == 8:
            return score, 35, 21, 4
        elif score == 9:
            return score, 16, 8, 0
        elif score == 10:
            return score, 0, 0, 0

def Rev_Kat_grouped(df):
    #Albumine: nu g/L, naar g/dL
    #CRP: nu mg/L, naar mg/dL
    #Calcium: nu mmol/L, naar mg/dL, 1 mmol/L = 4.01 mg/dl
    #Bili: nu umol, naar mg/dL, 1umol/l = 0.06 mg/dL
    #Platelets: <100.000/uL (zie artikel, staat fout op internet)
    # is gelijk aan <100 x10^9/L
    columns_to_check = ['Katagiri_Group', 'Visceral', 'Brain', 'Disseminated', 
                        "CRP", "LDH", "Alb", "Platelet", "Calc", "Tot_Bill",
                        "ECOG", "Pre_Chem", "Nr_Skel_Met"]
    
    missing_values = df[columns_to_check].isnull().any().any()
    
    if missing_values:
        return "MV", "MV", "MV"
    
    else:
        score = 0
        
        if df["Katagiri_Group"] == 2:
            score += 2
        if df["Katagiri_Group"] == 3:
            score += 3
            
        if df["Disseminated"] == 1:
            score += 2
        if (df["Disseminated"] == 0 and (df["Visceral"] == 1 or 
                                           df["Brain"] == 1)):
            score += 1
            
        if (df["Platelet"] < 100 or 
            df["Calc"]*4.01 >= 10.3 or 
            df["Tot_Bill"]*0.06 >= 1.4):
            score += 2
        if (df["Platelet"] >= 100 and 
              df["Calc"]*4.01 < 10.3 and
              df["Tot_Bill"]*0.06 < 1.4 and (df["CRP"]/10 >= 4 or 
                                               df["LDH"] > 250 or 
                                               df["Alb"]/10 < 3.7)):
            score += 1
            
        if df["ECOG"] in [3,4]:
            score += 1
            
        if df["Pre_Chem"] == 1:
            score += 1
            
        if df["Nr_Skel_Met"] == 1:
            score +=  1
            
        if score in [0, 1, 2, 3]:
            return score, 99, 98, 91
        elif score in [4, 5, 6]:
            return score, 84, 74, 49
        elif score in [7, 8, 9, 10]:
            return score, 43, 27, 6
```

```python
# Define a dictionary mapping function names to their respective column prefixes
scoring_systems = {
    Bollen: "Bollen",
    Mod_Bauer: "Mod_Bauer",
    Mizumoto: "Mizumoto",
    NESMS: "NESMS",
    Orig_Bauer: "Orig_Bauer",
    OSRI: "OSRI",
    Rev_Kat_individual: "Rev_Kat_individual",
    Rev_Kat_grouped: "Rev_Kat_grouped",
    Rev_Toku: "Rev_Toku",
    Tomita: "Tomita",
    van_der_Linden: "van_der_Linden"}

# Loop through the dictionary, applying each function and assigning results
for func, prefix in scoring_systems.items():
    df[[f"{prefix}_score", f"{prefix}_3_months", f"{prefix}_6_months", f"{prefix}_12_months"]] = df.apply(func, axis=1, result_type="expand")

df.to_excel("Data_scoring_results", index = False)
```
